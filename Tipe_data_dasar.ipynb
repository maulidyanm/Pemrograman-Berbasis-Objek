#Praktikum 01 - Tipe data dasar

```python
#integer
angka_int = 10
print("angka_int =", angka_int, "->", type(angka_int))

#float
angka_float = 3.14
print("Angka float =", angka_float, "->", type(angka_float))

#complex
angka_complex = 2 + 3j
print("Angka Kompleks =", angka_complex, "->", type(angka_complex))

#boolean
is_active = True
print("is_active =", is_active, "->", type(is_active))

#string
teks = "Hello, Python!"
print("teks =", teks, "->", type(teks))

#list
daftar_angka = [1, 2, 3, 4, 5]
print("daftar_angka =", daftar_angka, "->", type(daftar_angka))

#tuple
koordinat = (10, 20)
print("koordinat =", koordinat, "->", type(koordinat))

#dictionary
data_mahasiswa = {"nama": "Andi",
                  "nim": "All.2022.12345",
                  "jurusan": "Teknik Informatika"}
print("data_mahasiswa =", data_mahasiswa, "->", type(data_mahasiswa))

#set
himpunan_angka = {1, 2, 3, 2, 1}
print("himpunan_angka =", himpunan_angka, "->", type(himpunan_angka))

#contoh penggunaan konversi tipe data
nilai_str = "100"
print("\nnilai_str", nilai_str, "->", type(nilai_str))
nilai_int = int(nilai_str)
print("nilai_int =", nilai_int, "->", type(nilai_int))
```

#Praktikum 02 - Variabel dan Operasi Dasar

```python
#deklarasi variabel
nama = "Budi"
umur = 20
tinggi = 170.5
is_student = True

print("Nama =", nama)
print("Umur =", umur)
print("Tinggi =", tinggi, "cm")
print("Mahasiswa = ", is_student)

#operasi aritmetika
a = 10
b = 3

penjumlahan = a + b
pengurangan = a - b
perkalian = a * b
pembagian = a / b
pembagian_bulat = a // b
modulus = a % b
pangkat = a ** b

print("\nOPERASI ARITMETIKA")
print("a =", a, "b =", b)
print("Penjumlahan =", penjumlahan)
print("Pengurangan =", pengurangan)
print("Perkalian =", perkalian)
print("Pembagian =", pembagian)
print("Pembagian Bulat =", pembagian_bulat)
print("Modulus =", modulus)
print("Pangkat =", pangkat)

#Operasi Perbandingan
#Menghasilkan nilai Boolean (True atau False)
lebih_besar = a > b
kurang_dari = a < b
sama_dengan = a == b
tidak_sama = a != b
lebih_besar_sama = a >= b
kurang_sama = a <= b

print("\nOPERASI PERBANDINGAN")
print("a > b =", lebih_besar)
print("a < b =", kurang_dari)
print("a == b =", sama_dengan)
print("a != b =", tidak_sama)
print("a >= b =", lebih_besar_sama)
print("a <= b =", kurang_sama)

#Operasi Logika
x = True
y = False

logika_and = x and y
logika_or = x or y
logika_not_x = not x

print("\nOPERASI LOGIKA")
print("x =", x, "y =", y)
print("x and y =", logika_and)
print("x or y =", logika_or)
print("not x =", logika_not_x)

#penggunaan dalam percabangan
if a > b and b > 0:
    print("\nKondisi terpenuhi: a lebih besar dari b dan b masih positif")
else:
    print("\nKondisi tidak terpenuhi atau b <= 0")
```

#Praktikum 03 - Percabangan

```python
#if sederhana
nilai = 85
print("Contoh if sederhana")
if nilai >= 80:
    print("Selamat Anda lulus dengan nilai tingg.\n")

#if-else
umur = 17
print("Contoh if-else")
if umur >= 18:
    print("Anda sudah cukup umur untuk mendapatkan SIM.")
else:
    print("Anda belum cukup umur untuk mendapatkan SIM.\n")

#if-elif-else
hari = "Rabu"
print("Contoh if-elif-else")
if hari == "Senin":
    print("Hari Senin - Saatnya kembali bekerja!")
elif hari == "Selasa":
    print("Hari Selasa - Jadwal rapat mingguan.")
elif hari == "Rabu":
    print("Hari Rabu - Ada diskon di beberpa toko!")
else:
    print("Hari lainnya - Atur jadwalmu dengan baik.\n")

#if bersarang (nested if)
suhu = 35
print("\nContoh if bersarang (nested if)")
if suhu > 30:
    print("Cuaca cukup panas.")
    if suhu > 40:
        print("Bahkan sangat terik! Disarankan banyak minum air.")
    else:
        print("Masih relatif normal, tapi tetap jaga kesehatan.")
else:
    print("Cuaca sepertinya cukup sejuk.\n")

#menggabungkan operasi percabangan dengan operasi logika
nilai_teori = 75
nilai_praktik = 80
print("\nContoh IF dengan operasi logika AND/OR")
if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Anda lulus karena nilai teori dan praktik sudah memadai.")
elif nilai_teori < 70 or nilai_praktik < 70:
    print("Anda perlu meningkatkan nilai teori dan praktik.")
elif nilai_teori < 70:
    print("Anda perlu meningkatkan nilai teori.")
else:
    print("Anda perlu meningkatkan nilai praktik.\n")

#penggunaan if ternary
angka = -5
print("Contoh if ternary")
status = "Positif" if angka > 0 else "Negatif atau Nol"
print("Angka", angka, "adalah", status)
```

#Praktikum 04 - Perulangan

```python
#for loop dengan range
print("1) FOR loop dengan range")
for i in range(5):
    print("Perulangan ke-:", i)

print()

#for loop untuk mengiterasi list
print("2) FOR loop untuk mengiterasi list")
buah = ["apel", "mangga", "jeruk", "pisang"]
for item in buah:
    print("Buah:", item)

print()

#WHILE loop
print("3) WHILE loop sederhana")
count = 0
while count < 5:
    print("count:", count)
    count += 1

print()

#BREAK pada loop
print("4) BREAK pada loop")
for i in range(10):
    if i == 3:
        print("Loop dihentikan pada i =", i)
        break
    print("i:", i)

print()

#CONTINUE pada loop
print("5) CONTINUE pada loop")
for i in range(5):
    if i == 2:
        print("Lewati i =", i, "dengan continue")
        continue
    print("i:", i)

print()

#Nested loop (loop bersarang)
print("6) Nested loop (loop bersarang)")
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")

print()

#memanfaatkan else pada loop
print("7) Memanfaatkan else pada loop")
for x in range(3):
    print("x:", x)
else:
    print("Loop for telah selesai tanpa break.\n")

y = 0
while y < 3:
    print("y:", y)
    y += 1
else:
    print("Loop while telah selesai tanpa break.\n")

#PASS sebagai placeholder
print("8) PASS sebagai placeholder")
for i in range(3):
    if i == 1:
        pass  # Placeholder untuk kode yang akan ditambahkan nanti
    print("i:", i)
```

#Penugasan 01 - Body Mass Index (BMI)

```python
berat = float(input("Masukkan berat badan (kg): "))
tinggi_cm = float(input("Masukkan tinggi badan (cm): "))

tinggi_m = tinggi_cm / 100
bmi = berat /(tinggi_m ** 2)

#kategori
if bmi < 18.5:
  kategori = "Underweight"
elif 18.5 <= bmi <= 24.9:
  kategori = "Normal"
elif 25 <= bmi <= 29.9:
  kategori = "Overweight"
else:
  kategori = "Obese"

print("BMI Anda:", round(bmi, 2))
print("Kategri:", kategori)
```

#Penugasan 02 - Pengecekan Bilangan

```python
angka = int(input("Masukkan sebuah bilangan: "))

if angka % 2 == 0:
  print("Bilangan Genap")
else:
  print("BIlangan Ganjil")

if angka > 1:
  prima = True
  for i in range(2, angka):
    if angka % i == 0:
      prima = False
      break

  if prima:
    print("Bilangan Prima")
  else:
    print("Bukan Bilangan Prima")
else:
  print("Bukan Bilangan Prima")
```
