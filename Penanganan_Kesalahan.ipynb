#Praktikum 01 - try..except Dasar dalam Metode Kelas

```python
class DataStore:
  def __init__(self, nama_store="Default Store"):
    self._data ={}
    self.nama = nama_store
    print(f"DataStore '{self.nama}' dibuat")

  def tambah_data(self, key, value):
    print(f"[{self.nama}] Menambahkan data: '{key}' -> {value}")
    self._data[key] = value
    print("-> Data berhasil ditambahkan")

  def get_data(self, key):
    print(f"[{self.nama}] Mencoba mengambil data dengan key: '{key}'")
    try:
      nilai = self._data[key]
      print(f"-> Data ditemukan: {nilai}")
      return nilai
    except KeyError:
      print(f"-> ERROR: Key '{key}' tidak ditemukan di DataStore '{self.nama}'")
      return None

if __name__ == "__main__":
  store_produk = DataStore("Toko Produk")
  store_produk.tambah_data("Buku Tulis", 15000)
  store_produk.tambah_data("Pensil 2B", 2000)

  print("\nMengambil data yang ada:")
  harga_buku =store_produk.get_data("Buku Tulis")
  if harga_buku is not None:
    print(f"Harga buku tulis adalah: {harga_buku}")

  print("\nMengambil data yang tidak ada:")
  harga_penggaris = store_produk.get_data("Penggaris")
  if harga_penggaris is None:
    print(f"Harga penggaris tidak ditemukan (sesuai ekspektasi)")
```

#Praktikum 02 - Menangani Beberapa Exception Spesifik

```python
class KalkulatorAngka:
  def __init__ (self):
    print("Kalkulator Angka siap")

  def bagi(self, angka1, angka2):
    print(f"\nMencoba membagi '{angka1}' dengan '{angka2}'...")
    try:
      num1 = float(angka1)
      num2 = float(angka2)

      hasil = num1 / num2

      print(f"-> Hasil pembagian: {hasil}")
      return hasil
    except ValueError:
      print("-> ERROR: Input tidak valid! Pastika kedua input adalah angka")
      return None
    except ZeroDivisionError:
      print("-> ERROR: Tidak bisa melakukan pembagian dengan nol")
      return None
    except Exception as e:
      print(f"-> ERROR Lainnya: Terjadi kesalahan - {e}")
      return None

if __name__ == "__main__":
  calc = KalkulatorAngka()

  print("==== Tes Kalkulator ====")
  hasil1 = calc.bagi(100, 50)
  hasil2 = calc.bagi(20, 0)
  hasil3 = calc.bagi("limapuluh", 10)
  hasil4 = calc.bagi(50, "sepuluh")

  print("==== Ringkasan Hasil ====")
  print(f"Hasil 1 (100/50): {hasil1}")
  print(f"Hasil 2 (20/0): {hasil2}")
  print(f"Hasil 3 (limapuluh/10): {hasil3}")
  print(f"Hasil 4 (50/sepuluh): {hasil4}")
```

#Praktikum 03 - Menggunakan else dan finally dalam Metode

```python
import os

class FileProcessor:
  def __init__(self, nama_file):
    self.nama_file = nama_file
    self._file_handle = None
    print(f"\n-> FileProcessor dibuat untuk file: '{self.nama_file}'")

  def baca_dan_proses(self):
    print(f"[{self.nama_file}] Mencoba memproses...")
    try:
      print(f"[{self.nama_file}] Membuka file...")
      self._file_handle = open(self.nama_file, 'r', encoding='utf-8')

      print(f"[{self.nama_file}] Membaca isi file...")
      isi = self._file_handle.read()
      print(f"-> File berhasil dibuka")

      print(f"[{self.nama_file}] Memproses isi...")
      panjang_isi = len(isi)
      print(f"-> Panjang isi: {panjang_isi} karakter")

    except FileNotFoundError:
      print(f"-> ERROR: Gagal membaca '{self.nama_file}' tidak ditemukan")
    except UnicodeDecodeError:
      print(f"-> ERROR: Gagal membaca '{self.nama_file}' sebagai teks (mungkin file biner?)")
    except Exception as e:
      print(f"-> ERROR Lainnya saat memproses file: {type(e).__name__} - {e}")
    else:
      print(f"-> Pemrosesan Berhasil: Tidak ada error yang terjadi pada '{self.nama_file}'")
    finally:
      print(f"[{self.nama_file}] Blok Finally: Melakukan cleanup...")

      if self._file_handle and not self._file_handle.closed:
        print(f"Menutup file '{self.nama_file}'...")
        self._file_handle.close()
        print("-> File berhasil ditutup")
      else:
        print("-> File tidak perlu ditutup (belum dibuka atau sudah ditutup)")

    print(f"[{self.nama_file}] Selesai mencoba proses")
    print("-"*30)

if __name__ == "__main__":
  nama_file_sukses = "data_penting.txt"
  nama_file_gagal = "file_yang_tidak_ada.log"

  try:
    with open(nama_file_sukses, "w", encoding='utf-8') as f:
      f.write("Ini adalah baris pertama \n")
      f.write("Baris kedua berisi data penting \n")
    print(f"File dummy '{nama_file_sukses}' berhasil dibuat")
  except IOError as e:
    print(f"Gagal membuat file dummy '{nama_file_sukses}': {e}")

  print("\n--- Skenario 1: Proses Sukses ---")
  processor1 = FileProcessor(nama_file_sukses)
  processor1.baca_dan_proses()

  print("\n--- Skenario 1: File tidak ditemukan ---")
  processor2 = FileProcessor(nama_file_gagal)
  processor2.baca_dan_proses()

  try:
    if os.path.exists(nama_file_sukses):
      os.remove(nama_file_sukses)
      print(f"\nFile dummy '{nama_file_sukses}' dihapus")
  except OSError as e:
    print(f"\nGagal menghapus file dummy '{nama_file_sukses}': {e}")
```

#Praktikum 04 - Mengakses Objek Exception (as e)

```python
class KalkulatorAngka:
  def __init__(self):
    print("Kalkulator siap...")

  def bagi(self, angka1, angka2):
    print(f"\nMencoba membagi '{angka1}' dengan '{angka2}'...")
    try:
      num1 = float(angka1)
      num2 = float(angka2)
      if num2 == 0:
        raise ZeroDivisionError("Pembagi tidak boleh nol")

      hasil = num1 / num2
      print(f"-> Hasil Pembagian: {hasil}")
      return hasil
    except ValueError as e_val:
      print(f"-> ERROR Konversi: Input Tidak valid")
      print(f"Detail error (ValueError): {e_val}")
      return None
    except ZeroDivisionError as e_zero:
      print(f"-> ERROR Pembagian: Tidak dapat membagi dengan nol")
      print(f"Detail Error (ZeroDivisionError): {e_zero}")
      return None
    except Exception as e:
      print("Error Lainnya: terjadi kesalahan tidak terduga")
      print(f"Detail Error (Lainnya): {type(e).__name__} - {e}")
      return None

if __name__ == "__main__":
  calc = KalkulatorAngka()

  print("==== Tes Kalkulator ====")
  hasil1 = calc.bagi(100, 50)
  hasil2 = calc.bagi(20, 0)
  hasil3 = calc.bagi("limapuluh", 10)
  hasil4 = calc.bagi(50, "sepuluh")

  print("==== Ringkasan Hasil ====")
  print(f"Hasil 1 (100/50): {hasil1}")
  print(f"Hasil 2 (20/0): {hasil2}")
  print(f"Hasil 3 (limapuluh/10): {hasil3}")
  print(f"Hasil 4 (50/sepuluh): {hasil4}")
```

#Praktikum 05 - Membangkitkan Exception (raise) dalam Metode

```python
import locale
try:
  locale.setlocale(locale.LC_ALL, 'id_ID.UTF-8')
except locale.Error:
  print("Locale id_ID tidak tersedia")
def format_rupiah(angka):
    try:
        return locale.currency(angka, grouping=True, symbol='Rp ')
    except Exception:
        return f"Rp {angka:,.2f}"

class PengelolaAkun:
  def __init__(self, nomor_akun, saldo_awal=0):
    print(f"\nMencoba membuat akun {nomor_akun}...")
    if not isinstance(saldo_awal, (int, float)) or saldo_awal < 0:
      raise ValueError(f"Saldo awal {saldo_awal} tidak valid, harus angka non negatif")

    self.nomor_akun = nomor_akun
    self._saldo = float(saldo_awal)
    print(f"-> Akun {self.nomor_akun} berhasil dibuat dengan saldo: {format_rupiah(self._saldo)}")

  def get_saldo(self):
    print(f"[Info Akun {self.nomor_akun}] Saldo saat ini: {format_rupiah(self._saldo)}")
    return self._saldo

  def tarik_tunai(self, jumlah):
    print(f"\n[{self.nomor_akun}] Mencoba tarik tunai: {format_rupiah(jumlah)}")
    if not isinstance(jumlah, (int, float)) or jumlah <= 0:
      raise ValueError(f"Jumlah penarikan ({jumlah}) tidak valid, harus angka positif")

    if jumlah > self._saldo:
      raise RuntimeError(f"Saldo tidak mencukup! Saldo: {format_rupiah(self._saldo)}, Diminta: {format_rupiah(jumlah)}")

    self._saldo -= jumlah
    print(f"-> Penarikan {format_rupiah(jumlah)} berhasil")
    print(f"-> Saldo sekarang: {format_rupiah(self._saldo)}")
    return True

if __name__ == "__main__":
  print("\n==== Skenario 1: Saldo Awal Tidak Valid ====")
  try:
    akun_gagal = PengelolaAkun("ACC001", -50000)
  except ValueError as e:
    print(f"Error ditangkap saat pembuatan akun: {e}")

  print("\n==== Skenario 2: Akun Valid dan Transaksi ====")
  try:
    akun_sukses = PengelolaAkun("ACC002", 100000)
    akun_sukses.tarik_tunai(30000)
    print("\nMencoba tarik jumlah negatif...")
    akun_sukses.tarik_tunai(-10000)
  except ValueError as e:
    print(f"Error ditangkap (ValueError): {e}")
  except RuntimeError as e:
    print(f"Error ditangkap (RuntimeErroe): {e}")

  print("\n==== Skenario 3: Saldo Tidak Cukup ====")
  try:
    print(f"Saldo akun ACC02 saat ini: {format_rupiah(akun_sukses.get_saldo())}")
    print("Mencoba tarik jumlah > saldo....")
    akun_sukses.tarik_tunai(80000)
  except ValueError as e:
    print(f"Error ditangkap (ValueError): {e}")
  except RuntimeError as e:
    print(f"Error ditangkap (RuntimeError): {e}")

  print(f"\nSaldo akhir akun ACC002: {format_rupiah(akun_sukses.get_saldo())}")
```

#Praktikum 06 - Kombinasi Menangkap dan Membangkitkan Exception

```python
class KonfigurasiError(Exception):
  def __init__(self, message, penyebab_asli=None):
    super().__init__(message)
    self.penyebab_asli =penyebab_asli

  def __str__(self):
    if self.penyebab_asli:
      return f"{super().__str__()} (Disebabkan oleh: {type(self.penyebab_asli).__name__})"
      return super().__str__()

class PengelolaSetting:
  def __init__(self, file_config):
    self._config = {}
    self.file_config = file_config
    try:
      self._load_config()
      print(f"-> PengelolaSetting berhasil dibuat untuk '{self.file_config}'")
    except KonfigurasiError as e:
      print(f"-> Gagal membuat PengelolaSetting: {e}")
      self._config = {}
      print("-> Menggunakan konfigurasi default/kosong")

  def _load_config(self):
    print(f"\n[{self.file_config}] Mencoba load konfigurasi...")
    try:
      print(f"Membuka file '{self.file_config}'...")
      if self.file_config != 'config.txt':
        raise FileNotFoundError(f"File '{self.file_config}' tidak ada")

      print(f"Membaca dan validasi isi '{self.file_config}'...")
      data_raw = {"host": "192.168.1.1", "port": "8080"}
      data_raw = {"host": "192.168.1.1", "port": "xyz"}
      data_raw = {"host": "192.168.1.1"}

      port_str = data_raw['port']
      port_int = int(port_str)

      self.config['port'] = port_int
      self.config['host'] = data_raw.get('host', 'localhost')
      print("-> Konfigurasi berhasil di-load dan divalidasi")
    except FileNotFoundError as e_fnf:
      print(f"Error bawaan: File tidak ditemukan")
      raise KonfigurasiError(f"File konfigurasi penting '{self.file_config}' hilang", penyebab_asli=e_fnf) from e_fnf
    except KeyError as e_key:
      print(f"Error Bbawaan: Kunci konfigurasi hilang")
      raise KonfigurasiError(f"Kunci '{e_key}' wajib ada dalam file konfigurasi", penyebab_asli=e_key) from e_key
    except ValueError as e_val:
      print(f"Error bawaan: Nilai konfigurasi tidak valid")
      raise KonfigurasiError(f"Nilai untuk 'port' ('{port_str}') harus integer", penyebab_asli=e_val) from e_val
    except Exception as e_lain:
      print(f"Error bawaan tak terduga: {type(e_lain).__name__}")
      raise KonfigurasiError(f"Terjadi error tak terduga saat memuat konfigurasi", penyebab_asli=e_lain) from e_lain

  def get_setting(self, key, default=None):
    print(f"Mendapatkan setting '{key}'...")
    return self._config.get(key, default)

if __name__ == "__main__":
  print("\n==== Skenario 1: File salah ====")
  settings1 = PengelolaSetting("setting.cfg")
  print(f"Port dari settings1: {settings1.get_setting('port', 'Belum terload')}")

  print("\n==== Skenario 2: Data  port salah ====")
  settings2 = PengelolaSetting("config.txt")
  print(f"Port dari settings2: {settings2.get_setting('port', 'Belum terload')}")

  print("\n==== Skenario 3: Sukses ====")
  settings3 = PengelolaSetting("config.txt")
  print(f"Port dari settings3: {settings3.get_setting('port')}")
  print(f"Hasil dari settings3: {settings3.get_setting('host')}")
```

#Penugasan 01

```python
class InventarisBarang:
    def __init__(self):
        self._items = {}
        print("Inventaris barang berhasil dibuat")

    def tambah_barang(self, kode, nama, jumlah):
        try:
            if not isinstance(jumlah, int) or jumlah <= 0:
                raise ValueError("Jumlah harus integer positif")

            if kode in self._items:
                raise KeyError(f"Kode barang '{kode}' sudah ada")

            self._items[kode] = {
                "nama": nama,
                "stok": jumlah
            }

            print(f"Barang '{nama}' berhasil ditambahkan dengan stok {jumlah}")

        except ValueError as e:
            print(f"Error ValueError: {e}")

        except KeyError as e:
            print(f"Error KeyError: {e}")

    def ambil_barang(self, kode, jumlah):
        try:
            if not isinstance(jumlah, int) or jumlah <= 0:
                raise ValueError("Jumlah harus integer positif")

            if kode not in self._items:
                raise KeyError(f"Kode barang '{kode}' tidak ditemukan")

            if self._items[kode]["stok"] < jumlah:
                stok_tersedia = self._items[kode]["stok"]
                raise RuntimeError(
                    f"Stok barang '{kode}' tidak mencukupi "
                    f"(tersedia: {stok_tersedia}, diminta: {jumlah})"
                )

            self._items[kode]["stok"] -= jumlah

            print(f"Berhasil mengambil {jumlah} barang '{kode}'")
            print(f"Sisa stok: {self._items[kode]['stok']}")

        except ValueError as e:
            print(f"Error ValueError: {e}")

        except KeyError as e:
            print(f"Error KeyError: {e}")

        except RuntimeError as e:
            print(f"Error RuntimeError: {e}")

    def cek_stok(self, kode):
        try:
            if kode not in self._items:
                raise KeyError(f"Kode barang '{kode}' tidak ditemukan")

            barang = self._items[kode]

            print(f"Kode Barang : {kode}")
            print(f"Nama Barang : {barang['nama']}")
            print(f"Stok        : {barang['stok']}")

        except KeyError as e:
            print(f"Error KeyError: {e}")


if __name__ == "__main__":
    inventaris = InventarisBarang()

    print("\n=== Tambah Barang ===")
    inventaris.tambah_barang("BRG001", "Laptop", 10)
    inventaris.tambah_barang("BRG002", "Mouse", 20)

    print("\n=== Coba Tambah Kode Sama ===")
    inventaris.tambah_barang("BRG001", "Keyboard", 5)

    print("\n=== Cek Stok ===")
    inventaris.cek_stok("BRG001")

    print("\n=== Ambil Barang ===")
    inventaris.ambil_barang("BRG001", 3)

    print("\n=== Ambil Barang Melebihi Stok ===")
    inventaris.ambil_barang("BRG001", 20)

    print("\n=== Ambil Barang dengan Kode Salah ===")
    inventaris.ambil_barang("BRG999", 1)

    print("\n=== Cek Stok Setelah Pengambilan ===")
    inventaris.cek_stok("BRG001")

    print("\n=== Cek Stok Barang Tidak Ada ===")
    inventaris.cek_stok("BRG999")
```
