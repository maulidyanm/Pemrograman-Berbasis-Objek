#Praktikum 01 - Enkapsulasi pada kelas Bank Account

```python
class BankAccount:
  def __init__(self, owner, balance):
    self.__owner = owner
    self.__balance = balance

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f"{amount} telah ditambahkanke akun {self.__owner}.")
    else:
      print("Jumlah deposit harus lebih dari 0")

  def withdraw(self, amount):
    if amount <= self.__balance:
      self.__balance -= amount
      print(f"{amount} telah ditarik dari akun {self.__owner}")
    else:
      print("Saldo tiak mendukupi")

  def get_balance(self):
    return self.__balance

#penggunaan
if __name__ == "__main__":
  alice_account = BankAccount(owner = "Alice", balance=1000)

  #deposit
  alice_account.deposit(500)
  alice_account.deposit(-100)

  #withdraw
  alice_account.withdraw(300)
  alice_account.withdraw(2000)

  #dapatsaldo
  current_balance =alice_account.get_balance()
  print(f"Saldo terakhir di akun {alice_account._BankAccount__owner}: {current_balance}")
```

#Praktikum 02 - Enkapsulasi pada kelas Employee

```python
class Employee:
  def __init__(self, name, salary):
    self.__name = name
    self.__salary = salary

  def update_salary(self, increase):
    if increase > 0:
      self.__salary += increase
      print(f"Gaji telah dinaikan sebesar {increase}")
    else:
      print(f"Nilai kenaikan harus lebih dari 0")

  def set_salary(self, new_salary):
    if new_salary >= 0:
      self.__salary= new_salary
      print(f'Gaji diatur ulang menjadi {new_salary}')
    else:
      print(f"Gaji tidak dapat bernilai negatif")

  def get_salary(self):
    return self.__salary

  def get_employee_info(self):
    return f"Employee: {self.__name}, Gaji: {self.__salary}"


#penggunaan
if __name__ == "__main__":
  employee1 = Employee("John Doe", 50000)
  print(employee1.get_employee_info())

  employee1.update_salary(5000)
  print(f"Gaji setelah kenaikan: {employee1.get_salary()}")

  employee1.set_salary(60000)
  print(f"Informasi terbaru: {employee1.get_employee_info()}")
```

#Praktikum 03 - Konstruktor dan Destruktor Sederhana

```python
class SimpleExample:
  def __init__(self, name):
    self.name = name
    print(f"Konstruktor: Objek '{self.name}' telah dibuat")

  def __del__(self):
    print(f"Destruktor: Objek '{self.name}' telah dihapus")

def main():
  print("Program Dimulai\n")

  obj = SimpleExample("Demo")
  print("Program sedang berjalan....\n")

  del obj
  print("Objek telah dihapus secara eksplisit\n")

  print("Program Selesai")

if __name__ == "__main__":
  main()
```

#Praktikum 04 - Konstruktor dan Destruktor program FileLogger

```python
class FileLogger:
  def __init__(self, filename):
    self.filename = filename
    try:
      self.file = open(filename, "a")
      print(f"File '{self.filename}' berhasil dibuka untuk logging")
    except Exception as e:
      print(f"Gagal membuka file '{self.filename}': {e}")

  def write_log(self, message):
    self.file.write(message + "\n")
    self.file.flush()
    print(f"Pesan log: '{message} telah ditulis'")

  def __del__(self):
    if hasattr(self, "file") and not self.file.closed:
      self.file.close()
      print(f"File '{self.filename}' telah ditutup")

#penggunaan
if __name__ == "__main__":
  logger = FileLogger("application.log")

  logger.write_log("Aplikasi dimulai")
  logger.write_log("Melakukan operasi A...")
  logger.write_log("Operasi A selesai")
  logger.write_log("Aplikasi akan segera selesai")

  del logger
```

#Praktikum 05 - Properti "self" pada kelas

```python
class Calculator:
  def __init__(self, initial_value=0):
    self.value = initial_value
    print(f"Kalkulator diinisialisasi dengan nilai: {self.value}")

  def add(self, number):
    self.value += number
    print(f"Setelah penambahan {number}, nilai sekarang adalah: {self.value}")

  def subtract(self, number):
    self.value -=number
    print(f"Setelah pengurangan {number}, nilai sekarang adalah: {self.value}")

  def reset(self):
    self.value = 0
    print("Nilai telah direset ke 0")

  def show_value(self):
    print(f"Nilai saat ini adalah: {self.value}")

#penggunaan
def main():
  calc1 = Calculator(initial_value=10)

  calc1.add(5)
  calc1.subtract(3)
  calc1.show_value()

  calc2 = Calculator()
  calc2.add(20)
  calc2.show_value()

if __name__ == "__main__":
  main()
```

#Praktikum 06 - Menggunakan getter, setter, dan dekorator

```python
class Person:
  def __init__(self, name, age):
    self.__name = name
    self.__age = age

  @property
  def name(self):
    return self.__name

  @name.setter
  def name(self, value):
    if not value:
      print("Nama tidak boleh kosong")
    else:
      self.__name = value

  @property
  def age(self):
    return self.__age

  @age.setter
  def age(self, value):
    if value < 0:
      print("Umur tidak boleh negatif")
    else:
      self.__age = value

#penggunaan
def main():
  person = Person("Alice", 30)
  print(f"Nama: {person.name}, Umur: {person.age}")

  person.name = "Bob"
  person.age = 35
  print(f"Nama baru: {person.name}, Umur baru: {person.age}")

  person.age= -5

if __name__ == "__main__":
  main()
```

#Penugasan 01 - Grade

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    @property
    def score(self):
        return self._score

    @score.setter
    def score(self, value):
        if 0 <= value <= 100:
            self._score = value
            self._grade = self.hitung_grade(value)
        else:
            print("Nilai harus antara 0 - 100")

    @property
    def grade(self):
        return self._grade

    def hitung_grade(self, nilai):
        if nilai >= 90:
            return "A"
        elif nilai >= 80:
            return "B"
        elif nilai >= 70:
            return "C"
        elif nilai >= 60:
            return "D"
        else:
            return "E"

    def show_info(self):
        print(f"Nama Mahasiswa: {self.name}")
        print(f"Nilai: {self.score}")
        print(f"Grade: {self.grade}")

    def __del__(self):
        print(f"Data mahasiswa {self.name} telah dihapus dari sistem.")

nama = input("Masukkan nama: ")
nilai = int(input("Masukkan nilai: "))

mhs = Student(nama, nilai)

print("\nData awal:")
mhs.show_info()

# Ubah nilai
print("\nNilai diubah...")
nilai_baru = int(input("Masukkan nilai baru: "))
mhs.score = nilai_baru

print("\nData setelah diubah:")
mhs.show_info()

# Hapus objek
del mhs
```
