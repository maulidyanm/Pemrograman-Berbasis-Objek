#Praktikum 01 - Membuat dan Menggunakan Fungsi

```python
print("Ini adalah contoh fungsi built-in")

kata = "Pemrograman"
panjang_kata = len(kata)

angka = [10, 5, 30, 40, 25]
nilai_max = max(angka)
print(f"Nilai maksimum dalam daftar {angka} adalah: {nilai_max}")

def cetak_kuadrat(angka):
  print(f"Kuadrat dari angka{angka} adalah: {angka ** 2}")

def hitung_luas_persegi_panjang(panjang, lebar):
  return panjang * lebar

def info_mahasiswa(nama, umur, ipk):
  print(f"Nama: {nama}, Umur: {umur}, IPK: {ipk}")

def sapa_pengguna(nama):
  print(f"Halo, {nama}! Selamat datang di dunia Python!")

def hitung_keliling_persegi(sisi):
  return 4 * sisi

cetak_kuadrat(5)

luas = hitung_luas_persegi_panjang(10, 5)
print(f"Luas persegi panjang: {luas}")

info_mahasiswa("Budi", 22, 3.8)

sapa_pengguna("Andi")

keliling = hitung_keliling_persegi(5)
print(f"Keliling persegi dengan sisi 5 adalah: {keliling}")
```

#Praktikum 03 - Membuat Kelas Sederhana

```python
class Buku:
  def __init__(self, judul, pengarang, tahun_terbit):
    self.judul = judul
    self.pengarang = pengarang
    self.tahun_terbit = tahun_terbit
    self.status = "Tersedia"

  def tampilkan_info(self):
    print(f"Judul: {self.judul}")
    print(f"Pengarang: {self.pengarang}")
    print(f"Tahun terbit: {self.tahun_terbit}")
    print(f"Status: {self.status}")

  def pinjam(self):
    if self.status == "Tersedia":
      self.status = "Dipinjam"
      print(f"Buku '{self.judul}' telah dipinjam")
    else:
      print(f"Buku '{self.judul}' sedang dipinjam")

  def kembalikan(self):
    if self.status == "Dipinjam":
      self.status ="Tersedia"
      print(f"Buku '{self.judul}' telah dikembalikan")
    else:
      print(f"Buku '{self.judul}' tidak sedang dipinjam")


#objek
buku1 = Buku("Pemrograman Python", "John Doe", 2021)
buku2 = Buku("Data Sience untuk Pemula", "Jane Smith", 2020)

buku1.tampilkan_info()
buku2.pinjam()

buku2.kembalikan()
buku1.pinjam()
buku1.tampilkan_info()

#kelas mahasiswa
class Mahasiswa:
  def __init__(self, nama, nim, umur):
    self.nama = nama
    self.nim = nim
    self.umur = umur

  def tampilkan_info(self):
    print(f"Nama: {self.nama}")
    print(f"NIM: {self.nim}")
    print(f"Umur: {self.umur} tahun")

  def ubah_umur(self, umur_baru):
    self.umur = umur_baru

mahasiswa1 = Mahasiswa("Andi", "12345", 20)
mahasiswa2 = Mahasiswa("Budi", "67890", 22)

mahasiswa1.tampilkan_info()
print()
mahasiswa2.tampilkan_info()

mahasiswa1.ubah_umur(21)
print("\nSetelah mengubah umur mahasiswa1: ")
mahasiswa1.tampilkan_info()
```

#Praktikum 04 - Membuat dan Menggunakan Method dalam Kelas

```python
class Mobil:
  def __init__(self, merk, warna, tahun, harga):
    self.merk = merk
    self.warna = warna
    self.tahun = tahun
    self.harga = harga

  def tampilkan_info(self):
    print(f"Mobil {self.merk} berwarna {self.warna}, tahun {self.tahun}, harga: Rp{self.harga},00")

  def diskon(self, persen_diskon):
    diskon_harga = self.harga * (persen_diskon / 100)
    harga_setelah_diskon = self.harga - diskon_harga
    print(f"Harga setelah diskon {persen_diskon}%: Rp{harga_setelah_diskon},00")

  def hitung_usia(self, tahun_sekarang):
    usia = tahun_sekarang - self.tahun
    return usia

  def perbarui_harga(self, harga_baru, tahun_baru):
    self.harga = harga_baru
    self.tahun = tahun_baru
    print(f"Harga dan tahun mobil {self.merk} diperbarui menjadi Rp{self.harga},00 dan tahun {self.tahun}")

#objek mobil
mobil1 = Mobil("Toyota", "Hitam", 2015, 300000000)
mobil2 = Mobil("Honda", "Merah", 2018, 250000000)

mobil1.tampilkan_info()
mobil2.tampilkan_info()

mobil1.diskon(10)
mobil2.diskon(15)

usia_mobil1 = mobil1.hitung_usia(2025)
usia_mobil2 = mobil2.hitung_usia(2025)

print(f"Usia mobil1 pada tahun 2025: {usia_mobil1} tahun")
print(f"Usia mobil2 pada tahun 2025: {usia_mobil2} tahun")

mobil1.perbarui_harga(280000000, 2022)
mobil2.perbarui_harga(240000000, 2021)
```

#Penugasan 01 - Parkir Roda Dua

```python
class ParkirMotor:
    def __init__(self, plat, jam_masuk):
        self.plat = plat
        self.jam_masuk = jam_masuk
        self.jam_keluar = 0
        self.biaya = 0

    def keluar(self, jam_keluar):
        self.jam_keluar = jam_keluar
        lama = self.jam_keluar - self.jam_masuk

        if lama <= 0:
            lama = 1

        self.biaya = lama * 2000
        return self.biaya

data_parkir = []

def masuk():
    plat = input("Masukkan plat nomor: ")
    jam = float(input("Jam masuk: "))

    motor = ParkirMotor(plat, jam)
    data_parkir.append(motor)
    print("Motor masuk parkir")

def keluar():
    plat = input("Masukkan plat nomor: ")
    for motor in data_parkir:
        if motor.plat == plat:
            jam_keluar = int(input("Jam keluar: "))
            biaya = motor.keluar(jam_keluar)

            print("Biaya parkir:", biaya)
            data_parkir.remove(motor)
            return

    print("Data tidak ditemukan!")

def tampilkan():
    if len(data_parkir) == 0:
        print("Tidak ada kendaraan parkir")
    else:
        for motor in data_parkir:
            print(f"Plat: {motor.plat}, Jam Masuk: {motor.jam_masuk}")


def menu():
    while True:
        print("\n=== PARKIR MOTOR ===")
        print("1. Motor Masuk")
        print("2. Motor Keluar")
        print("3. Tampilkan Data")
        print("4. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            masuk()
        elif pilihan == "2":
            keluar()
        elif pilihan == "3":
            tampilkan()
        elif pilihan == "4":
            print("Program selesai")
            break
        else:
            print("Pilihan tidak valid!")
menu()
```
